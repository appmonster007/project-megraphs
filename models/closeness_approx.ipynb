```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
# from sklearn. import 
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('closeness_M_web-edu.csv')
df_y = df['closeness']
df_inp = df.drop(['closeness','Nodes','betweenness'],axis=1)
```

```python
corr = df.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
TEST_SHARE = 0.95
Train_Input, Test_Input, Train_Label, Test_Label = train_test_split(df_inp,df_y,test_size=TEST_SHARE)

model = RandomForestRegressor()
# print(df.__sizeof__(),Train_Input.__sizeof__(),Test_Input.__sizeof__())
```

```python
model.fit(Train_Input,Train_Label)
```

```python
prediction_labels = model.predict(Test_Input)
compare_df = pd.DataFrame({'actual':Test_Label,'predict':prediction_labels})
```

```python

# plt.scatter(list(compare_df.index.to_numpy()),list(compare_df['actual'].to_numpy()))
# plt.scatter(list(compare_df.index.to_numpy()),list(compare_df['predict'].to_numpy()))
# print(compare_df.index.to_numpy().size,compare_df['actual'].to_numpy().size)
# plt.show()
```

```python
fig, ax = plt.subplots(figsize=(16, 8))
x = list(compare_df.index.to_numpy())
actual = list(compare_df['actual'].to_numpy())
predict = list(compare_df['predict'].to_numpy())

ax.scatter(x,actual,color='blue')
ax.scatter(x,predict,color='orange')

# ax.set_ylim([0,0.3])
plt.show()
```

```python
print(model.score(Test_Input,Test_Label),model.score(Train_Input,Train_Label))
```
